```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import random as nr
np.set_printoptions(precision=2)
```

```python
n = 10000
r1 = nr.randn(1,n)
r1.shape
```

```python
x = np.arange(1,n+1)
y = r1[0]
T = np.arctan2(y,1)
```

```python
plt.axes([0.5,0.5,0.95,0.95])
plt.scatter(x,y, s=5, c=T, alpha=.9)
plt.xlim(0,1000), plt.xticks(np.linspace(0,10000,5))
plt.ylim(-4,4), plt.yticks(np.linspace(-4,4,5))
plt.show()
```

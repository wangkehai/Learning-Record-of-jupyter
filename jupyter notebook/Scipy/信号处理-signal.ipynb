Scipy库的signal模块提供了信号处理方面的许多函数，包括卷积运算，B样条，滤波及滤波器设计等。

### 中值滤波

中值滤波能够有效地消除声音信号中的瞬间噪声或者图像中的斑点噪声。在signal模块中，medfilt()用于对一维信号进行中值滤波，medfilt2d()用于对二维信号进行中值滤波。在scipy.ndimage模块中另有针对多维图像的中值滤波器。以下介绍medfilt()的效果。

```python
from scipy import signal
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0,20,0.1)
x = np.sin(t)
x[np.random.randint(0,len(t),20)] += np.random.standard_normal(20)*0.6
x2 = signal.medfilt(x,5)
x3 = signal.order_filter(x,np.ones(5),2)
print(np.all(x2 == x3))

fig = plt.figure(figsize=(8,8), dpi=150)
plt.plot(t,x,label="before")
plt.plot(t,x2+0.5,alpha=0.6, label="after")
plt.legend(loc="best")
plt.show
```

**signal.medfilt(data,kernel_size)**参数共有两个，第一个参数是要处理的信号源，第二个参数是计算中值的窗口大小，即滤波窗口的size，它必须为奇数。medfilt()将信号中的每个值都替换为其窗口内的中值。

中值滤波是排序滤波的一种，使用排序滤波可将元素替换为其窗格内制定排序顺序的元素，其调用形式如下：

**order_filter(a,domain,rank)**

其中a是多维数组，domain是和a维数相同的数组，rank是一个非负整数，用来选择窗口中元素排序后的值，0表示最小值，1为第二小的值。中值滤波也可用order_filter()计算，上面例子数组x3中domain参数为一个长度为5、值全为1的数组。
